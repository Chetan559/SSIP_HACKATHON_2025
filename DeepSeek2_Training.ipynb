```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import bitsandbytes
tokenizer = AutoTokenizer.from_pretrained("brijesh12/deepseek-7b-chat-SSIP2")
model = AutoModelForCausalLM.from_pretrained(
    "brijesh12/deepseek-7b-chat-SSIP2",
    torch_dtype="auto",
    device_map="auto"
).to("cuda")


```

```python
import json

# Load MOSDAC dataset
with open("mosdac_mission_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Function to generate Q&A pairs
def generate_qna(text, title):
    questions = [
        ("What is {}?".format(title), text[:100]),  # Basic question
        ("What is the purpose of {}?".format(title), " ".join(text.split()[:10])),  # Purpose-based question
        ("Who operates {}?".format(title), "ISRO")  # Generic question
    ]

    qas = []
    for idx, (question, answer) in enumerate(questions):
        answer_start = text.find(answer)
        if answer_start != -1:  # Ensure answer exists in text
            qas.append({
                "question": question,
                "id": f"{title}-{idx}",
                "answers": [{"text": answer, "answer_start": answer_start}]
            })

    return {"context": text, "qas": qas}

# Convert dataset
qa_data = {"data": []}
for title, content in data.items():
    if "text" in content:  # Extract text content
        qa_data["data"].append({
            "title": title,
            "paragraphs": [generate_qna(content["text"], title)]
        })

# Save as JSON
with open("qa_dataset.json", "w", encoding="utf-8") as f:
    json.dump(qa_data, f, indent=4)

print("✅ Q&A Dataset saved as 'qa_dataset.json'.")
```

```python
import json
from datasets import Dataset

# Load dataset
with open("mosdac_mission_data.json", "r") as f:
    raw_data = json.load(f)

# Prepare Q&A pairs
qa_pairs = []
for key, value in raw_data.items():
    if "text" in value:
        question = f"What is {key.replace('-', ' ')}?"
        answer = value["text"]
        qa_pairs.append({"question": question, "answer": answer})

# Convert to Hugging Face dataset format
dataset = Dataset.from_list(qa_pairs)

# Check dataset structure
print(dataset[0])  # Should output {'question': 'What is INSAT 3D?', 'answer': 'INSAT-3D is a...'}


def tokenize_function(examples):
    # Tokenize the input questions
    model_inputs = tokenizer(
        examples["question"],
        truncation=True,
        padding="max_length",
        max_length=256,
        return_tensors="pt"
    )

    # Tokenize the output answers
    labels = tokenizer(
        examples["answer"],
        truncation=True,
        padding="max_length",
        max_length=256,
        return_tensors="pt"
    )

    # Assign tokenized labels
    model_inputs["labels"] = labels["input_ids"]

    return model_inputs

# Apply tokenization
tokenized_datasets = dataset.map(tokenize_function, batched=True, remove_columns=["question", "answer"])

# Check dataset structure
print(tokenized_datasets.column_names)  # Should show ['input_ids', 'attention_mask', 'labels']


# Check structure after tokenization
print(tokenized_datasets.column_names)  # Should show ['input_ids', 'attention_mask', 'labels']

```

```python
from peft import LoraConfig, get_peft_model
from transformers import TrainingArguments, Trainer

# Define LoRA Config
lora_config = LoraConfig(
    r=16, lora_alpha=16, lora_dropout=0.05, bias="none",
    task_type="CAUSAL_LM"
)



# Apply LoRA
model = get_peft_model(model, lora_config)

# Training arguments (No evaluation)
training_args = TrainingArguments(
    output_dir="./deepseek-finetuned",
    evaluation_strategy="no",  # No evaluation
    learning_rate=2e-5,
    per_device_train_batch_size=2,  # Adjust based on GPU memory
    num_train_epochs=4,
    weight_decay=0.01,
    save_strategy="epoch",
    remove_unused_columns=False,
    gradient_accumulation_steps=8,  # Helps reduce memory usage
    fp16=True  # Enables mixed precision training (reduces memory usage)
)

# Trainer (No eval dataset)
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets,  # Only training dataset
    processing_class=tokenizer
)

# Start fine-tuning
trainer.train()
```

```python
question = "what is insat-3dr?"
inputs = tokenizer(question, return_tensors="pt").to("cuda")
outputs = model.generate(**inputs, max_length=400)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
